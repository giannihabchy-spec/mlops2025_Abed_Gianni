```python
def convert_datetime(data, cols):
    for c in cols:
        data[c] = pd.to_datetime(data[c], errors="coerce")
    return data

data = convert_datetime(train, ["pickup_datetime", "dropoff_datetime"])

print(data.dtypes)
print(data)
```

```python
print("Rows:", len(df))
print("Total missing values:", df.isna().sum().sum())
print("Duplicate rows:", df.duplicated().sum())

if "trip_duration" in df.columns:
    print("Trip_duration < 60:", (df["trip_duration"] < 60).sum())
    print("Trip_duration > 10800:", (df["trip_duration"] > 10800).sum())

loc_cols = ["pickup_latitude","dropoff_latitude","pickup_longitude","dropoff_longitude"]
if all(c in df.columns for c in loc_cols):
    same_loc = ((df["pickup_latitude"] == df["dropoff_latitude"]) &
                (df["pickup_longitude"] == df["dropoff_longitude"])).sum()
    print("Same pickup/dropoff:", same_loc)
```
